## object detection

```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
model = load_model('ankit_face_5k.h5')
```

```python
# Load the pre-trained face detection cascade classifier from OpenCV
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# Open a connection to the webcam (you can replace this with a video file if needed)
cap = cv2.VideoCapture(0)

while True:
    # Read a frame from the webcam
    ret, frame = cap.read()

    if not ret:
        break

    # Convert the frame to grayscale for face detection
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces in the frame
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in faces:
        # Extract the face region
        im = frame[y:y+h, x:x+w]
        im = frame
        im = PIL.Image.fromarray(im)
        im = np.array(im)
        cv2.imwrite(r"C:\Users\ANKIT SINGH\Desktop\Face_recognition\check.jpg", im)
        im = PIL.Image.open(r"C:\Users\ANKIT SINGH\Desktop\Face_recognition\check.jpg").convert("L")
        im = im.resize((162,162))
        im = np.array(im)
        im = np.reshape(im,(-1,162,162,1))
        pred = model.predict(im, steps = 1)
        if pred >= 0.5:
            result = 'ANKIT'
            color = (0, 0, 255)
        else:
            result = 'No_Anomaly'
            color = (0, 255, 0)
        # Draw a rectangle around the detected face
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv2.putText(frame, str(result), (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    # Display the frame
    cv2.imshow('Face Id', frame)

    # Exit the loop when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close the OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
